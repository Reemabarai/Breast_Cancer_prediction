```python
#MISSING VALUE
print("\n Missing value :")
breast.isnull().sum()
```

```python
breast.drop(['Unnamed: 32'], axis=1, inplace=True, errors='ignore')
```

```python
breast.shape
```

```python
#HANDLING DULICATION
duplicated_raws = breast.duplicated().sum()
print(f"Number of duplicated raws:",duplicated_raws)
```

```python
breast['diagnosis'].value_counts()
```

```python
#ENCODING CATEGORICAL DATA
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
breast['diagnosis'] = le.fit_transform(breast['diagnosis'])
print("categorical data encoded succefully")
breast.head()
```

```python
breast['diagnosis'].value_counts()
```

```python
#FEATURE SCALING
from sklearn.preprocessing import StandardScaler

feature = breast.drop('diagnosis',axis=1)
target=breast['diagnosis']
sc = StandardScaler()
scaled_features = sc.fit_transform(feature)
print("Feature scaling complete")

```

```python
x = breast.drop('diagnosis',axis=1)
y = breast['diagnosis']
```

```python
x.shape
```

```python
# Apply StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(feature)
# Convert back to DataFrame
breast_scaled = pd.DataFrame(scaled_features, columns=feature.columns)
print(breast_scaled.head())
```

```python
#splitting data into traning and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(breast.iloc[:,1:10].corr(),annot=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
rf_pred = rf_model.predict(x_test)

rf_acc = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_acc)
```

```python
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(x_train, y_train)
y_pred_svc = svc_classifier.predict(x_test)
svm_acc = accuracy_score(y_test, y_pred_svc)
print("SVM Accuracy (Linear Kernel):", svm_acc)
```

```python
from sklearn.metrics import classification_report

# Random Forest Report
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_pred))

# SVM Report
print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svc))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Calculate accuracy
rf_acc = accuracy_score(y_test, rf_pred)
svm_acc = accuracy_score(y_test, y_pred_svc)

# Plot
plt.bar(['Random Forest', 'SVM'], [rf_acc, svm_acc], color=['skyblue', 'salmon'])
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```
